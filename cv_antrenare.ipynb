```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/Resume.csv")
```

```python
df
```

```python
df=df.drop('ID',axis=1)
df
```

```python
df=df.drop('Resume_html',axis=1)
df
```

```python
import re
import string

def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    return text
```

```python
df["Resume_str"] = df["Resume_str"].apply(clean_text)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score


X_train, X_test, y_train, y_test = train_test_split(df["Resume_str"], df["Category"], test_size=0.2, random_state=42)
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, "save_model.pkl")
```

```python
def classify_resume(cv_text):
    cv_text = clean_text(cv_text)
    model = joblib.load("resume_classifier.pkl")
    prediction = model.predict([cv_text])
    return prediction[0]
```

```python
import PyPDF2

# Deschide fișierul PDF
pdfFileObj = open("CV.pdf", "rb")

# Creează un obiect PdfReader
pdfReader = PyPDF2.PdfReader(pdfFileObj)

# Numărul de pagini
num_pages = len(pdfReader.pages)

# Extrage textul din fiecare pagină
text = ""
for page in range(num_pages):
    text += pdfReader.pages[page].extract_text()

pdfFileObj.close()

print(text)
```

```python
print("Predicted Job:", classify_resume(text))
```
